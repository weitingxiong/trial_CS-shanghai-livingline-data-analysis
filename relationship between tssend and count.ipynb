```python
import pandas as pd
import json
import matplotlib.pyplot as plt
import io
import re
import datetime
import pytz
import urllib
from io import open
from pandas import Series
```

```python
filepath='E:/0 MIT_CS/CS_Livingline/03 analysis files/files for analysis'
with open(filepath+'./2018-06-19_00','r',encoding='utf-8',errors='replace') as datafile:
    data=datafile.readlines()
    dataRawJson=[json.loads(data[i]) for i in range(len(data))]
df = pd.DataFrame.from_records(dataRawJson)
df
```

```python
wifidata=df['wifidata']
df1 = pd.DataFrame.from_records(wifidata)
df1
```

```python
ts_col=df1['tssend']
df1['realtime'] = df1['tssend']
new_time = []
new_time_s = []
timeCorrection=8*60*60
for record in ts_col:
    each_ts=float(record)+timeCorrection
#     allTimes.extend([ts])
    dateTime=datetime.datetime.utcfromtimestamp(each_ts/1000).strftime('%Y%m%d %H:%M:%S')
    new_time.append(dateTime)
    new_time_s.append(long(each_ts/1000/60/60))
#     print(dateTime)
df1['realtime']=Series(new_time, index = df1.index)
df1['tssecond']=Series(new_time_s, index = df1.index)
df1
# ts().dateTime()
```
